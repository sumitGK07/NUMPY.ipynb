Q9. (Combined - Random + Reshape + Statistics)
Generate a 1D array of 20 random integers between 1 and 50 using
np.random.randint().
Reshape it into a 4x5 matrix.
Print the sum, mean, and standard deviation of the matrix.
Find the maximum value in each row.

```python
import numpy as np

# Generate 20 random integers between 1 and 50
arr = np.random.randint(1, 51, 20)

print("Original 1D Array:")
print(arr)

# Reshape into 4x5 matrix
matrix = arr.reshape(4, 5)

print("\n4x5 Matrix:")
print(matrix)

print("\nSum =", np.sum(matrix))
print("Mean =", np.mean(matrix))
print("Standard Deviation =", np.std(matrix))

print("\nMaximum value in each row:")
print(np.max(matrix, axis=1))
```

Q10. (Mini Project - NumPy Application)
Create a Simple Statistics Calculator using NumPy:
Write a program that:
Asks the user how many numbers they want to generate.
Generates that many random numbers between 10 and 100 using
np.random.randint().
Stores them in a 1D array.
Prints:
The array
Mean, Median, Standard Deviation
Minimum and Maximum
Reshapes into a 2D array (if possible) and prints row-wise sum.
Use proper comments and exception handling where needed.

```python
import numpy as np

try:
    n = int(input("How many numbers do you want to generate? "))

    if n <= 0:
        print("Please enter a positive number.")

    else:
        arr = np.random.randint(10, 101, n)

        print("\nGenerated Array:")
        print(arr)


        print("\nStatistics")
        print("Mean =", np.mean(arr))
        print("Median =", np.median(arr))
        print("Standard Deviation =", np.std(arr))
        print("Minimum =", np.min(arr))
        print("Maximum =", np.max(arr))

        rows = int(np.sqrt(n))

        while rows > 1:
            if n % rows == 0:
                cols = n // rows

                matrix = arr.reshape(rows, cols)

                print("\nReshaped 2D Array:")
                print(matrix)

                print("\nRow-wise Sum:")
                print(np.sum(matrix, axis=1))
                break

            rows -= 1
        else:
            print("\nReshaping into a 2D array is not possible.")

except ValueError:
    print("Invalid input! Please enter an integer.")
except Exception as e:
    print("Error:", e)
```
